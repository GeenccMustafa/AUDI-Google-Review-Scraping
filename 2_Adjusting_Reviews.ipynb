```python
import pandas as pd
import numpy as np
import re
```

```python
#read file
df = pd.read_excel("audi_rew.xlsx")
df
```

```python
#drop nan values
df['Review'] = df['Review'].astype(str)
df = df[df['Review'] != 'nan']

#reset index
df.reset_index(drop=True, inplace=True)
df
```

```python
def isProblematic(s):
    s = str(s)
    if len(s) == 0:
        return True
    elif s.endswith('‚Ä¶'):
        return True
    else:
        return False
```

```python
df['Problematic'] = df['Review'].apply(isProblematic)
df
```

```python
df['Problematic'].value_counts()
```

```python
def translated(x, lang = 'de'):
    if lang == 'de':
        if '√úbersetzt von Google' in x:
            return True
        else:
            return False
    elif lang == 'en':
        if 'Translated by Google' in x:
            return True
        else:
            return False
```

```python
#check which reviews are translated by google
df['Translated'] = df['Review'].apply(translated, lang = 'en')
df
```

```python
def adjustReview(r, lang = 'de'):
    if lang == 'de':
        if '(√úbersetzt von Google)' in r:
            r = r.split('(√úbersetzt von Google) ')[-1]

            if '(Original)' in r:
                r = r.split('\n\n(Original)')[0]
    elif lang == 'en':
        if '(Translated by Google)' in r:
            r = r.split('(Translated by Google) ')[-1]

            if '(Original)' in r:
                r = r.split('\n\n(Original)')[0]
        
    
    return r
```

```python
df['Review'] = df['Review'].apply(adjustReview, lang = 'en')
df
```

```python
def extractYear(s, lang = 'de'):
    year = 2022
    
    if lang == 'de':
        try:
            _, num, unit = s.split(' ')

            if (num == 'einem') or (num == 'einer'):
                num = 1
            else:
                num = int(num)

            if unit == 'Monaten':
                if num > 4:
                    year = 2021
            elif 'Jahr' in unit:
                year = year - num

        except:
            pass
    elif lang == 'en':
        try:
            num, unit, _ = s.split(' ')

            if (num == 'a') or (num == 'one'):
                num = 1
            else:
                num = int(num)

            if unit == 'months':
                if num > 4:
                    year = 2021
            elif 'year' in unit:
                year = year - num

        except:
            pass
        
        
    return year
```

```python
df['Year'] = df['Date'].apply(extractYear, lang = 'en')
df
```

```python
def adjustImageUrl(url):
    url = url[:url.find('=')+1] + 'w224-h224'
    return url
```

```python
df['Image URL'] = df['Image URL'].apply(adjustImageUrl)
```

```python
def isRealPicture(url):
    url = url[len('https://lh3.googleusercontent.com/'):]
    key = url.split('/')[0]
    result = False
    if key == 'a-':
        result = True
    
    return result  
```

```python
df['Profile picture'] = df['Image URL'].apply(isRealPicture)
```

```python
df['Profile picture'].value_counts()
```

```python
df = df[df['Problematic'] == False]
df.drop(columns = ['Problematic'], inplace = True)
```

```python
df.reset_index(drop = True, inplace = True)
df
```
